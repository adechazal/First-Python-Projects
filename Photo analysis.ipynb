# Import a photo
## playing around with a photo using numpy

```python
import numpy as np
from skimage import io
photo = io.imread('surf.jpg')
type(photo)
```

```python
#Have a look at what has been imported: 4912 row by 7360 columns and 3 RGB channels
photo.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(photo)
```

```python
#reverses the order of the columns. So, you get a reverse the image. Note the array is start, stop, step in the array
plt.imshow(photo[::-1])
```

```python
#returns all of the rows and reversed the columns so it inverses the image
plt.imshow(photo[:, ::-1])
```

```python
#crop the photo
plt.imshow(photo[2000:3000, 3000:5000])
```

```python
#halve the size of the image by taking every other row and every other column
plt.imshow(photo[::2, ::2])
```

```python
#take the sin of every element in the np array (not useful for a photo but instructive)
photo
photo_sin = np.sin(photo)
photo_sin
```

```python
print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.max(photo))
print(np.min(photo))
print(np.argmax(photo))
print(np.argmin(photo))
print(np.median(photo))
```

```python
#mask the photo. where the value in the photo is greater than 100 replace it with 255 and where it is less than 100 leave it (replace with nothing)
photo_masked = np.where(photo > 100,255,0)
plt.figure(figsize=(18, 10))
plt.imshow(photo_masked[:, 1000:7500])
plt.savefig('surf1.jpg')
```

```python
#transpose an array so that the columns are rows and vice versa (landscape to portrait)
plt.imshow(photo[:,:,0].T)
```
